```python
#Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import data from csv files
city_df = pd.read_csv('raw_data/city_data.csv')
ride_df = pd.read_csv('raw_data/ride_data.csv')
```

```python
#Merge data from both files based on city
merged_df = pd.merge(ride_df,city_df,how = "left",on="city")
```

```python
#Separate data based on type - Urban, Suburban and Rural
urban = merged_df[merged_df["type"] == "Urban"]
suburban = merged_df[merged_df["type"] == "Suburban"]
rural = merged_df[merged_df["type"] == "Rural"]
```

```python
#Check sample data
urban.head()
```

```python
#Find Average fare, ride count and drivers per city for all areas
#Urban
urbanaveragefare = urban.groupby("city").mean()["fare"]
urbanridecount = urban.groupby("city").count()["ride_id"]
urbandriverspercity = urban.groupby("city").mean()["driver_count"]
#Suburban
suburbanaveragefare = suburban.groupby("city").mean()["fare"]
suburbanridecount = suburban.groupby("city").count()["ride_id"]
suburbandriverspercity = suburban.groupby("city").mean()["driver_count"]
#Rural
ruralaveragefare = rural.groupby("city").mean()["fare"]
ruralridecount = rural.groupby("city").count()["ride_id"]
ruraldriverspercity = rural.groupby("city").mean()["driver_count"]
```

```python
# Create a Scatter plot containing all 3 areas and present them in different colors
plt.scatter(urbanridecount,urbanaveragefare,s=10*urbandriverspercity,
            color="red",  marker='o',
            edgecolor="black", linewidths=1, alpha=0.8, label="Urban")

plt.scatter(suburbanridecount,suburbanaveragefare,s=10*suburbandriverspercity,
            color="blue",  marker='o',
            edgecolor="black", linewidths=1, alpha=0.8, label="Suburban")

plt.scatter(ruralridecount,ruralaveragefare,s=10*ruraldriverspercity,
            color="yellow",  marker='o',
            edgecolor="black", linewidths=1, alpha=0.8, label="Rural")
#Add labels, grid, title, legend and define limit for x axis
plt.title("Pyber Analysis")
plt.ylabel("Average Fare")
plt.xlabel("Ride Count")
plt.legend(loc="best")
plt.xlim(0,36)
plt.grid(True)
plt.text(37,35,"Analysis:\n \
- Average fare is higher in Urban areas, followed by Suburban and Rural")
#Show plot
plt.show()
```

```python
#Calculate the percentage fare by area
percent_fare_by_type = merged_df.groupby("type").sum()["fare"] / merged_df["fare"].sum() * 100

percent_fare_by_type
#define title
plt.title("Pyber - Percentage Fare by Area")
#create pie plot
plt.pie(percent_fare_by_type,
        labels=["Rural","Suburban","Urban"],
        colors=["yellow","blue","red"],autopct='%1.1f%%',
        explode=[0,0,0.1],
       shadow= True,
       startangle=150)
plt.text(2,1,"Analysis:\n \
- Average fare is higher in Urban areas, followed by Suburban and Rural")
#show plot
plt.show()
```

```python
#Calculate the percentage of rides by area
percent_rides_by_type = merged_df.groupby("type").count()["ride_id"] / merged_df["ride_id"].count() * 100

percent_rides_by_type
#define title
plt.title("Pyber - Percentage Rides per Area")
#create pie plot
plt.pie(percent_rides_by_type,
        labels=["Rural","Suburban","Urban"],
        colors=["yellow","blue","red"],autopct='%1.1f%%',
        explode=[0,0,0.1],
       shadow= True,
       startangle=150)
plt.text(2,1,"Analysis:\n \
- The number of rides is higher in Urban areas, followed by Suburban and Rural")
#display plot
plt.show()
```

```python
#Calculate the percentage of drivers by area
percent_drivers_by_type = merged_df.groupby("type").mean()["driver_count"] / merged_df["driver_count"].mean() * 100

percent_drivers_by_type
#define title
plt.title("Pyber - Percentage Drivers per Area")
#create pie plot
plt.pie(percent_drivers_by_type,
        labels=["Rural","Suburban","Urban"],
        colors=["yellow","blue","red"],autopct='%1.1f%%',
        explode=[0,0,0.1],
       shadow= True,
       startangle=150)
plt.text(2,1,"Analysis:\n \
- The number of drivers is higher in Urban areas, followed by Suburban and Rural")
#display plot
plt.show()
```
