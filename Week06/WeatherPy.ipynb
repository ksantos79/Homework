```python
# Dependencies
import openweathermapy.core as owm
from citipy import citipy
import random as rd
# Import api_key
from config import api_key
```

```python
lat = -180
lng = -90
cities = []
countrycodes = []
finallat = []
finallon = []
for x in range(0,1500):
    lat = rd.randint(-180,180)
    lng = rd.randint(-90,90)
    
    citydata = citipy.nearest_city(lng,lat)
    if citydata.city_name not in cities:
        cities.append(citydata.city_name)
        countrycodes.append(citydata.country_code)
        finallat.append(lat)
        finallon.append(lng)
print(len(cities))
```

```python
import requests
citieslat = []
citieslng = []
citiesname = []
countrycode = []
urllist = []
cityidlist = []
templist = []
humlist = []
cloudiness = []
windspeedlist = []
baseurl = "https://api.openweathermap.org/data/2.5/weather?q="
settings = {"units":"imperial","APIID":api_key}

for x in range(len(cities)):
    try:
        url = f"{baseurl}{cities[x]},{countrycodes[x]}&APPID={api_key}"
        print(f"Retrieving data for {cities[x]}, {countrycodes[x]}")
        print(url)
        print("")
        urllist.append(url)
        citydata = requests.get(url).json()
        citieslat.append(citydata['coord']['lat'])
        citieslng.append(citydata['coord']['lon'])
        cityidlist.append(citydata['id'])
        citiesname.append(cities[x])
        countrycode.append(countrycodes[x])
        cloudiness.append(citydata['clouds']['all'])
        windspeedlist.append(citydata['wind']['speed'])
        templist.append(citydata['main']['temp'])
        humlist.append(citydata['main']['humidity'])
    except:
        pass
    #print(citydata)
```

```python
len(citieslat)
```

```python
import pandas as pd

counter = 0
for x in templist:
    templist[counter]= x * 9/5 - 459.67
    counter += 1

cities_df = pd.DataFrame({"City ID":cityidlist,"City Name":citiesname,"Country Code":countrycode,
                          "Latitude":citieslat,"Longitude":citieslng,"Temperature":templist,"Humidity":humlist,
                          "Cloudiness":cloudiness,"Wind Speed":windspeedlist})



cities_df = cities_df[["City ID","City Name","Country Code","Latitude","Longitude","Temperature",
                      "Humidity","Cloudiness","Wind Speed"]]
cities_df.head()
```

```python
import matplotlib.pyplot as plt
x_axis = cities_df["Latitude"]
y_axis = cities_df["Temperature"]
plt.grid(True)    
plt.title("City Latitude vs Temperature (06/08/18)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.xlim(-80,100)
plt.ylim(0,110)
plt.scatter(x_axis,y_axis,  marker="o", color="blue")
plt.savefig("LatitudeVsTemperature.png")
plt.show()
```
